## 1. Project Overview

Financial institutions such as banks, fintech lenders, and credit providers face an important business challenge when issuing loans: they need to approve profitable borrowers while reducing the risk of lending to customers who may default.

This project develops a machine learning solution to predict whether a loan applicant is likely to default. The goal is not only to build a high-performing classification model, but also to evaluate the model from a business-risk perspective.

In credit risk modeling, model evaluation must go beyond accuracy. A model that looks accurate overall may still fail to detect risky borrowers. Therefore, this project focuses on recall, precision, ROC-AUC, precision-recall trade-offs, threshold selection, and business cost impact.

---

### 1.1 Business Problem

The core business problem is:

> How can a financial institution identify high-risk loan applicants before approval, while avoiding unnecessary rejection of potentially good customers?

Incorrect lending decisions can create two types of business risk:

1. **False Negative**
   - The model predicts that a borrower will not default, but the borrower actually defaults.
   - This is the most expensive error because the institution may approve a risky loan.

2. **False Positive**
   - The model predicts that a borrower will default, but the borrower would actually repay.
   - This creates lost revenue opportunity and may reduce customer acquisition.

Because false negatives are usually more costly in credit-risk settings, this project treats loan default prediction as a **cost-sensitive classification problem**.

### 1.3 Machine Learning Objective

The machine learning objective is to build a binary classification model that predicts the probability of loan default using borrower, loan, and credit-history features.

The model should:

- Predict default probability for each loan applicant

Because this is a credit-risk problem, the model is evaluated using multiple metrics:

| Metric | Why It Matters |
|---|---|

The final model will be selected based on both predictive performance and business usefulness.

### 2.6 Target Variable Distribution and Class Imbalance

The target variable for this project is `loan_status`.

* `0` = No Default
* `1` = Default

Before modeling, it is important to inspect the class distribution. In credit-risk datasets, the number of non-default cases is often higher than the number of default cases. This creates a class imbalance problem.

Class imbalance matters because a model can achieve high accuracy by predicting the majority class too often, while still failing to identify risky borrowers. Therefore, this project will evaluate the model using recall, precision, F1-score, ROC-AUC, precision-recall analysis, and business cost impact rather than accuracy alone.


```python
# ============================================================
# Target Variable Distribution
# ============================================================

target_col = "loan_status"

# Count and percentage distribution
target_counts = df[target_col].value_counts().sort_index()
target_percentages = (df[target_col].value_counts(normalize=True).sort_index() * 100).round(2)

target_distribution = pd.DataFrame({
    "loan_status": target_counts.index,
    "class_label": ["No Default", "Default"],
    "count": target_counts.values,
    "percentage": target_percentages.values
})

display(target_distribution)

# Plot class distribution
plt.figure(figsize=(7, 5))
sns.barplot(
    data=target_distribution,
    x="class_label",
    y="count"
)

plt.title("Loan Status Distribution")
plt.xlabel("Loan Status")
plt.ylabel("Number of Applications")
plt.show()

# Print business interpretation
default_rate = target_percentages.loc[1]
non_default_rate = target_percentages.loc[0]

print(f"Non-default applications: {non_default_rate:.2f}%")
print(f"Default applications: {default_rate:.2f}%")

if default_rate < 30:
    print("\nInterpretation:")
    print("The dataset is imbalanced because default cases represent a minority of the observations.")
    print("Accuracy alone may be misleading, so recall, precision, F1-score, ROC-AUC, and PR-AUC will be used for model evaluation.")
```

### 3.2 Categorical Risk Analysis

Categorical variables are important in credit-risk modeling because they often represent business-relevant borrower segments.

In this project, we analyze default rates across:

* `person_home_ownership`
* `loan_intent`
* `loan_grade`
* `cb_person_default_on_file`

These features help answer questions such as:

* Do renters default more often than homeowners?
* Are some loan purposes riskier than others?
* Does loan grade strongly separate low-risk and high-risk borrowers?
* Does previous default history increase future default risk?

The analysis below compares default rates across categorical groups and supports business interpretation before modeling.


```python
# ============================================================
# Default Rate by Categorical Features
# ============================================================

categorical_features = [
    "person_home_ownership",
    "loan_intent",
    "loan_grade",
    "cb_person_default_on_file"
]

def categorical_default_summary(data, feature, target="loan_status"):
    """
    Create a default-rate summary table for a categorical feature.
    """
    summary = (
        data.groupby(feature)
        .agg(
            applications=(target, "count"),
            defaults=(target, "sum"),
            default_rate=(target, "mean")
        )
        .reset_index()
    )

    summary["default_rate"] = (summary["default_rate"] * 100).round(2)
    summary = summary.sort_values("default_rate", ascending=False)

    return summary


categorical_summary_tables = {}

for feature in categorical_features:
    print(f"\nDefault Risk Summary by {feature}")
    summary_table = categorical_default_summary(df, feature)
    categorical_summary_tables[feature] = summary_table
    display(summary_table)

    plt.figure(figsize=(8, 5))
    sns.barplot(
        data=summary_table,
        x=feature,
        y="default_rate"
    )

    plt.title(f"Default Rate by {feature}")
    plt.xlabel(feature)
    plt.ylabel("Default Rate (%)")
    plt.xticks(rotation=35, ha="right")
    plt.tight_layout()
    plt.show()
```

### 4.4 Missing Value Treatment Strategy

The missing-value assessment shows that only two columns contain missing values:

* `loan_int_rate`
* `person_emp_length`

Rows with missing values will not be removed because deleting them would reduce the dataset size and may remove useful borrower information.

The treatment strategy is:

1. **Create missing-value indicator features**

   * `loan_int_rate_missing`
   * `person_emp_length_missing`

   These indicators preserve information about whether a value was originally missing. This is useful because missingness itself may contain business information.

2. **Impute `person_emp_length` using the median**

   * Employment length is a numerical variable with outliers.
   * The median is more robust than the mean.

3. **Impute `loan_int_rate` using the median interest rate within each `loan_grade`**

   * Interest rate is closely related to loan grade.
   * Grade-based imputation is more business-relevant than using one global median for all borrowers.

4. **Use global median as a fallback**

   * If any loan grade has no available median interest rate, the overall median interest rate will be used as a fallback.

This strategy preserves the dataset, limits distortion from outliers, and uses domain logic where appropriate.


```python
# ============================================================
# Missing Value Treatment
# ============================================================

# Create a cleaned working dataset
df_clean = df.copy()

# Preserve missingness information before imputation
df_clean["loan_int_rate_missing"] = df_clean["loan_int_rate"].isna().astype(int)
df_clean["person_emp_length_missing"] = df_clean["person_emp_length"].isna().astype(int)

# Store missing values before treatment
missing_before = df_clean.isna().sum()

# 1. Impute person_emp_length using median
emp_length_median = df_clean["person_emp_length"].median()
df_clean["person_emp_length"] = df_clean["person_emp_length"].fillna(emp_length_median)

# 2. Impute loan_int_rate using loan_grade-specific median
grade_interest_medians = df_clean.groupby("loan_grade")["loan_int_rate"].transform("median")
global_interest_median = df_clean["loan_int_rate"].median()

df_clean["loan_int_rate"] = df_clean["loan_int_rate"].fillna(grade_interest_medians)
df_clean["loan_int_rate"] = df_clean["loan_int_rate"].fillna(global_interest_median)

# Store missing values after treatment
missing_after = df_clean.isna().sum()

# Create missing-value treatment log
missing_treatment_log = pd.DataFrame({
    "missing_before": missing_before,
    "missing_after": missing_after,
    "missing_reduced": missing_before - missing_after
})

missing_treatment_log = missing_treatment_log[missing_treatment_log["missing_before"] > 0]

display(missing_treatment_log)

print("Missing value treatment completed.")
print(f"Employment length median used: {emp_length_median:.2f}")
print(f"Global interest rate median fallback: {global_interest_median:.2f}")
print(f"Remaining missing values in df_clean: {df_clean.isna().sum().sum()}")
print(f"Dataset shape after missing value treatment: {df_clean.shape}")
```

### 4.5 Outlier Policy and Data Quality Corrections

The EDA and numerical summaries showed several unrealistic or extreme values that require careful treatment before modeling.

The outlier strategy is based on business logic and data-quality principles:

1. **Age**
   - Loan applicants should be adults.
   - Very high ages such as values above 80 are treated as unrealistic for this dataset.
   - Rows with `person_age > 80` are removed.

2. **Employment length**
   - Employment length cannot realistically exceed the applicant's adult working years.
   - Employment length is capped using the smaller value of:
     - 50 years
     - `person_age - 18`

3. **Income**
   - Extremely high income values can distort model training.
   - `person_income` is capped at the 99th percentile.
   - This preserves most observations while reducing the influence of extreme synthetic or data-entry values.

4. **Loan percent income**
   - Because `loan_percent_income` depends on income, it is recalculated after income capping.
   - This keeps the relationship between loan amount and income internally consistent.

The goal is not to remove risky borrowers. The goal is to reduce the influence of unrealistic values and improve model reliability.

```python
# ============================================================
# Outlier Policy and Impact Check
# ============================================================

# Work from the missing-value-treated dataset
df_model = df_clean.copy()

# Store before-cleaning information
shape_before_outliers = df_model.shape
summary_before_outliers = df_model[[
    "person_age",
    "person_income",
    "person_emp_length",
    "loan_amnt",
    "loan_percent_income"
]].describe().T

# ------------------------------------------------------------
# 1. Remove unrealistic age values
# ------------------------------------------------------------

age_filter = (df_model["person_age"] >= 18) & (df_model["person_age"] <= 80)
rows_before_age_filter = df_model.shape[0]

df_model = df_model[age_filter].copy()

rows_after_age_filter = df_model.shape[0]
age_rows_removed = rows_before_age_filter - rows_after_age_filter

# ------------------------------------------------------------
# 2. Cap employment length using business logic
# ------------------------------------------------------------

df_model["max_plausible_emp_length"] = (df_model["person_age"] - 18).clip(lower=0)

df_model["person_emp_length"] = np.minimum(
    df_model["person_emp_length"],
    df_model["max_plausible_emp_length"]
)

df_model["person_emp_length"] = np.minimum(
    df_model["person_emp_length"],
    50
)

# Remove helper column after correction
df_model = df_model.drop(columns=["max_plausible_emp_length"])

# ------------------------------------------------------------
# 3. Cap extreme income values at the 99th percentile
# ------------------------------------------------------------

income_cap = df_model["person_income"].quantile(0.99)

df_model["person_income"] = np.where(
    df_model["person_income"] > income_cap,
    income_cap,
    df_model["person_income"]
)

# ------------------------------------------------------------
# 4. Recalculate loan_percent_income after income capping
# ------------------------------------------------------------

df_model["loan_percent_income"] = df_model["loan_amnt"] / df_model["person_income"]
df_model["loan_percent_income"] = df_model["loan_percent_income"].round(4)

# Store after-cleaning information
shape_after_outliers = df_model.shape
summary_after_outliers = df_model[[
    "person_age",
    "person_income",
    "person_emp_length",
    "loan_amnt",
    "loan_percent_income"
]].describe().T

# ------------------------------------------------------------
# 5. Create outlier treatment log
# ------------------------------------------------------------

outlier_treatment_log = pd.DataFrame({
    "step": [
        "Initial dataset after missing-value treatment",
        "Remove unrealistic ages outside 18–80",
        "Cap employment length using age-based rule and 50-year maximum",
        "Cap income at 99th percentile",
        "Recalculate loan_percent_income"
    ],
    "rows_before": [
        shape_before_outliers[0],
        rows_before_age_filter,
        rows_after_age_filter,
        rows_after_age_filter,
        rows_after_age_filter
    ],
    "rows_after": [
        shape_before_outliers[0],
        rows_after_age_filter,
        rows_after_age_filter,
        rows_after_age_filter,
        rows_after_age_filter
    ],
    "rows_removed": [
        0,
        age_rows_removed,
        0,
        0,
        0
    ]
})

display(outlier_treatment_log)

print("Outlier treatment completed.")
print(f"Rows before outlier treatment: {shape_before_outliers[0]:,}")
print(f"Rows after outlier treatment: {shape_after_outliers[0]:,}")
print(f"Rows removed due to unrealistic age values: {age_rows_removed:,}")
print(f"Income cap used at 99th percentile: {income_cap:,.2f}")
print(f"Final modeling dataset shape: {df_model.shape}")

print("\nSummary before outlier treatment:")
display(summary_before_outliers)

print("\nSummary after outlier treatment:")
display(summary_after_outliers)
```

### 5.3 Preprocessing Pipeline Design

The dataset contains both numerical and categorical variables, so preprocessing must be handled carefully before modeling.

Two preprocessing pipelines are defined:

1. **Linear-model preprocessing**

   * Numerical features are scaled using `StandardScaler`.
   * Categorical features are encoded using one-hot encoding.
   * This pipeline is appropriate for Logistic Regression.

2. **Tree-model preprocessing**

   * Numerical features are passed through without scaling.
   * Categorical features are encoded using one-hot encoding.
   * This pipeline is appropriate for Random Forest and other tree-based models.

The preprocessing objects are defined at this stage but not fitted on the full dataset. They will be fitted later inside model pipelines after the train-test split. This avoids data leakage and keeps the workflow reproducible.


```python
# ============================================================
# Preprocessing Pipeline Definitions
# ============================================================

def make_onehot_encoder():
    """
    Create a OneHotEncoder with compatibility across scikit-learn versions.
    Newer versions use sparse_output=False, older versions use sparse=False.
    """
    try:
        return OneHotEncoder(handle_unknown="ignore", sparse_output=False)
    except TypeError:
        return OneHotEncoder(handle_unknown="ignore", sparse=False)


# Preprocessing for Logistic Regression
# Numerical features are scaled; categorical features are one-hot encoded.
preprocessor_linear = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), numerical_cols),
        ("cat", make_onehot_encoder(), categorical_cols)
    ],
    remainder="drop"
)


# Preprocessing for Random Forest
# Numerical features are not scaled; categorical features are one-hot encoded.
preprocessor_tree = ColumnTransformer(
    transformers=[
        ("num", "passthrough", numerical_cols),
        ("cat", make_onehot_encoder(), categorical_cols)
    ],
    remainder="drop"
)


# Summarize categorical cardinality and expected encoded feature count
categorical_cardinality = pd.DataFrame({
    "categorical_feature": categorical_cols,
    "unique_categories": [X[col].nunique() for col in categorical_cols]
})

display(categorical_cardinality)

estimated_encoded_feature_count = len(numerical_cols) + categorical_cardinality["unique_categories"].sum()

preprocessing_summary = pd.DataFrame({
    "pipeline": ["preprocessor_linear", "preprocessor_tree"],
    "intended_model_type": ["Logistic Regression", "Random Forest / Tree-Based Models"],
    "numerical_treatment": ["StandardScaler", "Passthrough"],
    "categorical_treatment": ["OneHotEncoder", "OneHotEncoder"],
    "number_of_numerical_features": [len(numerical_cols), len(numerical_cols)],
    "number_of_categorical_features": [len(categorical_cols), len(categorical_cols)],
    "estimated_features_after_encoding": [
        estimated_encoded_feature_count,
        estimated_encoded_feature_count
    ]
})

display(preprocessing_summary)

print("Preprocessing pipelines defined successfully.")
print(f"Estimated number of features after one-hot encoding: {estimated_encoded_feature_count}")
print("Important: preprocessors are defined but not fitted yet. They will be fitted inside model pipelines after train-test split.")
```

### 6.4 Reusable Model Evaluation Function

Before training models, a reusable evaluation function is created to ensure that all models are assessed using the same metrics.

The function evaluates:

- Accuracy
- Precision
- Recall
- F1-score
- ROC-AUC
- PR-AUC
- Confusion matrix values
- Classification report

This is important because the project compares multiple models and later tests different decision thresholds.

Using one evaluation function makes the notebook more consistent, reproducible, and easier to interpret.

```python
# ============================================================
# Reusable Model Evaluation Function
# ============================================================

def evaluate_classification_model(
    model,
    X_data,
    y_true,
    model_name="Model",
    threshold=0.50,
    show_confusion_matrix=True
):
    """
    Evaluate a binary classification model using probability predictions.

    Parameters:
    - model: fitted scikit-learn pipeline or classifier
    - X_data: feature matrix
    - y_true: true labels
    - model_name: name used in output tables
    - threshold: decision threshold for converting probabilities into class labels
    - show_confusion_matrix: whether to display a confusion matrix plot

    Returns:
    - metrics_dict: dictionary with key evaluation metrics
    """

    # Predicted probabilities for the positive class
    y_proba = model.predict_proba(X_data)[:, 1]

    # Convert probabilities into class predictions using selected threshold
    y_pred = (y_proba >= threshold).astype(int)

    # Confusion matrix values
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()

    # Evaluation metrics
    metrics_dict = {
        "model": model_name,
        "threshold": threshold,
        "accuracy": accuracy_score(y_true, y_pred),
        "precision": precision_score(y_true, y_pred, zero_division=0),
        "recall": recall_score(y_true, y_pred, zero_division=0),
        "f1_score": f1_score(y_true, y_pred, zero_division=0),
        "roc_auc": roc_auc_score(y_true, y_proba),
        "pr_auc": average_precision_score(y_true, y_proba),
        "true_negatives": tn,
        "false_positives": fp,
        "false_negatives": fn,
        "true_positives": tp
    }

    # Display metrics table
    metrics_table = pd.DataFrame([metrics_dict])
    display(metrics_table)

    # Display classification report
    print(f"Classification Report: {model_name}")
    print(classification_report(
        y_true,
        y_pred,
        target_names=["No Default", "Default"],
        zero_division=0
    ))

    # Display confusion matrix
    if show_confusion_matrix:
        disp = ConfusionMatrixDisplay(
            confusion_matrix=confusion_matrix(y_true, y_pred),
            display_labels=["No Default", "Default"]
        )
        disp.plot(values_format="d")
        plt.title(f"Confusion Matrix — {model_name} | Threshold = {threshold}")
        plt.show()

    return metrics_dict


print("Reusable evaluation function created successfully.")
```

```python
# ============================================================
# Cross-Validation: Logistic Regression
# ============================================================

# Define stratified cross-validation strategy
cv_strategy = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state=RANDOM_STATE
)

# Define scoring metrics
scoring_metrics = {
    "accuracy": "accuracy",
    "precision": "precision",
    "recall": "recall",
    "f1_score": "f1",
    "roc_auc": "roc_auc",
    "pr_auc": "average_precision"
}

# Run cross-validation on the training data
log_reg_cv_results = cross_validate(
    estimator=log_reg_model,
    X=X_train,
    y=y_train,
    cv=cv_strategy,
    scoring=scoring_metrics,
    n_jobs=-1,
    return_train_score=False
)

# Convert fold-level results into a clean table
log_reg_cv_table = pd.DataFrame({
    metric: log_reg_cv_results[f"test_{metric}"]
    for metric in scoring_metrics.keys()
})

log_reg_cv_table.index = [f"Fold {i}" for i in range(1, len(log_reg_cv_table) + 1)]

display(log_reg_cv_table.round(4))

# Create summary table
log_reg_cv_summary = pd.DataFrame({
    "metric": log_reg_cv_table.columns,
    "mean_score": log_reg_cv_table.mean().values,
    "std_score": log_reg_cv_table.std().values,
    "min_score": log_reg_cv_table.min().values,
    "max_score": log_reg_cv_table.max().values
})

display(log_reg_cv_summary.round(4))

print("Logistic Regression cross-validation completed successfully.")
print("The mean score shows average validation performance across folds.")
print("The standard deviation shows performance stability across folds.")
```

### 6.7 Threshold Analysis: Logistic Regression

By default, binary classification models use a probability threshold of 0.50.

However, in credit-risk modeling, the default threshold is not always the best business decision. A lower threshold may identify more potential defaulters, increasing recall, but it may also create more false positives by flagging more safe borrowers as risky.

This section compares several decision thresholds for the Logistic Regression baseline model.

The purpose is to understand the trade-off between:

* **False negatives**: risky borrowers incorrectly predicted as safe
* **False positives**: safe borrowers incorrectly predicted as risky
* **Recall**: ability to detect actual defaulters
* **Precision**: reliability of default predictions
* **F1-score**: balance between precision and recall

For this project, reducing false negatives is especially important because missed defaulters can create financial losses. However, precision also matters because rejecting too many safe borrowers can reduce business growth.


```python
# ============================================================
# Threshold Analysis: Logistic Regression
# ============================================================

# Predicted probabilities for the default class
log_reg_test_proba = log_reg_model.predict_proba(X_test)[:, 1]

# Candidate thresholds to compare
thresholds = [0.20, 0.25, 0.30, 0.35, 0.40, 0.45, 0.50]

threshold_results = []

for threshold in thresholds:
    y_pred_threshold = (log_reg_test_proba >= threshold).astype(int)

    tn, fp, fn, tp = confusion_matrix(y_test, y_pred_threshold).ravel()

    threshold_results.append({
        "threshold": threshold,
        "accuracy": accuracy_score(y_test, y_pred_threshold),
        "precision": precision_score(y_test, y_pred_threshold, zero_division=0),
        "recall": recall_score(y_test, y_pred_threshold, zero_division=0),
        "f1_score": f1_score(y_test, y_pred_threshold, zero_division=0),
        "true_negatives": tn,
        "false_positives": fp,
        "false_negatives": fn,
        "true_positives": tp
    })

threshold_results_df = pd.DataFrame(threshold_results)

display(threshold_results_df.round(4))

# Plot threshold trade-offs
plt.figure(figsize=(8, 5))
plt.plot(threshold_results_df["threshold"], threshold_results_df["precision"], marker="o", label="Precision")
plt.plot(threshold_results_df["threshold"], threshold_results_df["recall"], marker="o", label="Recall")
plt.plot(threshold_results_df["threshold"], threshold_results_df["f1_score"], marker="o", label="F1-score")

plt.title("Logistic Regression Threshold Trade-Off")
plt.xlabel("Decision Threshold")
plt.ylabel("Score")
plt.legend()
plt.tight_layout()
plt.show()

# Identify threshold with highest F1-score
best_f1_row = threshold_results_df.loc[threshold_results_df["f1_score"].idxmax()]

print("Best threshold by F1-score:")
display(pd.DataFrame([best_f1_row]).round(4))

print("Business interpretation:")
print("- Lower thresholds usually increase recall and reduce false negatives.")
print("- Higher thresholds usually increase precision and reduce false positives.")
print("- The final threshold should be selected based on business cost, not only statistical performance.")
```

### 6.11 Hyperparameter Tuning: Random Forest

The Random Forest baseline model already performs strongly, but hyperparameter tuning is used to check whether performance can be improved further.

Hyperparameter tuning helps optimize model settings such as:

* Number of trees
* Maximum tree depth
* Minimum samples required in leaf nodes
* Number of features considered at each split

Because this is a credit-risk classification problem, the tuning process should not focus only on accuracy. The model should balance default detection with precision.

In this step, `GridSearchCV` is used with stratified 5-fold cross-validation. The tuning process evaluates multiple metrics, but the best model is selected using F1-score because F1-score balances precision and recall.

The tuned Random Forest model will later be evaluated on the holdout test set and compared with the baseline Random Forest.


```python
# ============================================================
# Hyperparameter Tuning: Random Forest
# ============================================================

# Define Random Forest pipeline for tuning
rf_tuning_pipeline = Pipeline(steps=[
    ("preprocessor", preprocessor_tree),
    ("model", RandomForestClassifier(
        class_weight="balanced_subsample",
        random_state=RANDOM_STATE,
        n_jobs=-1
    ))
])

# Define a compact but meaningful parameter grid
rf_param_grid = {
    "model__n_estimators": [100, 200],
    "model__max_depth": [10, 15, None],
    "model__min_samples_leaf": [1, 2, 5],
    "model__max_features": ["sqrt"]
}

# Grid search with multiple scoring metrics
rf_grid_search = GridSearchCV(
    estimator=rf_tuning_pipeline,
    param_grid=rf_param_grid,
    scoring=scoring_metrics,
    refit="f1_score",
    cv=cv_strategy,
    n_jobs=1,
    verbose=1,
    return_train_score=True
)

# Fit grid search on training data
rf_grid_search.fit(X_train, y_train)

print("Random Forest hyperparameter tuning completed successfully.")
print(f"Best cross-validation F1-score: {rf_grid_search.best_score_:.4f}")
print("Best parameters:")
print(rf_grid_search.best_params_)

# Convert tuning results into a readable table
rf_tuning_results = pd.DataFrame(rf_grid_search.cv_results_)

selected_tuning_columns = [
    "mean_test_accuracy",
    "mean_test_precision",
    "mean_test_recall",
    "mean_test_f1_score",
    "mean_test_roc_auc",
    "mean_test_pr_auc",
    "std_test_f1_score",
    "param_model__n_estimators",
    "param_model__max_depth",
    "param_model__min_samples_leaf",
    "param_model__max_features"
]

rf_tuning_summary = (
    rf_tuning_results[selected_tuning_columns]
    .sort_values(by="mean_test_f1_score", ascending=False)
    .head(10)
)

display(rf_tuning_summary.round(4))

# Store best tuned model
rf_tuned_model = rf_grid_search.best_estimator_

# Evaluate tuned model on test set
rf_tuned_metrics = evaluate_classification_model(
    model=rf_tuned_model,
    X_data=X_test,
    y_true=y_test,
    model_name="Random Forest Tuned",
    threshold=0.50
)

# Store tuned model result without duplicating if the cell is rerun
if "model_results" not in globals():
    model_results = []

model_results = [
    result for result in model_results
    if result["model"] != "Random Forest Tuned"
]

model_results.append(rf_tuned_metrics)
```

### 6.14 Random Forest Threshold Analysis

The previous model comparison evaluated Random Forest models using the default threshold of 0.50.

However, credit-risk decisions often require a threshold that reflects business risk. A lower threshold may detect more actual defaulters, reducing false negatives, but it may also increase false positives by flagging more safe borrowers as risky.

This section compares several decision thresholds for both Random Forest models:

1. Random Forest Baseline
2. Random Forest Tuned

The goal is to evaluate whether threshold adjustment can improve the business usefulness of the Random Forest models.

The analysis focuses on:

* Precision
* Recall
* F1-score
* False positives
* False negatives
* True positives
* True negatives

This prepares the project for the next step: business cost simulation.


```python
# ============================================================
# Random Forest Threshold Analysis
# ============================================================

def threshold_analysis(model, X_data, y_true, model_name, thresholds):
    """
    Evaluate a fitted model across multiple classification thresholds.
    """
    y_proba = model.predict_proba(X_data)[:, 1]
    results = []

    for threshold in thresholds:
        y_pred = (y_proba >= threshold).astype(int)
        tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()

        results.append({
            "model": model_name,
            "threshold": threshold,
            "accuracy": accuracy_score(y_true, y_pred),
            "precision": precision_score(y_true, y_pred, zero_division=0),
            "recall": recall_score(y_true, y_pred, zero_division=0),
            "f1_score": f1_score(y_true, y_pred, zero_division=0),
            "false_positives": fp,
            "false_negatives": fn,
            "true_positives": tp,
            "true_negatives": tn
        })

    return pd.DataFrame(results)


# Candidate thresholds
rf_thresholds = [0.20, 0.25, 0.30, 0.35, 0.40, 0.45, 0.50]

# Run threshold analysis for both Random Forest models
rf_baseline_threshold_df = threshold_analysis(
    model=rf_model,
    X_data=X_test,
    y_true=y_test,
    model_name="Random Forest Baseline",
    thresholds=rf_thresholds
)

rf_tuned_threshold_df = threshold_analysis(
    model=rf_tuned_model,
    X_data=X_test,
    y_true=y_test,
    model_name="Random Forest Tuned",
    thresholds=rf_thresholds
)

# Combine results
rf_threshold_comparison_df = pd.concat(
    [rf_baseline_threshold_df, rf_tuned_threshold_df],
    ignore_index=True
)

display(rf_threshold_comparison_df.round(4))

# Best threshold by F1-score for each model
best_rf_thresholds_by_f1 = (
    rf_threshold_comparison_df
    .loc[rf_threshold_comparison_df.groupby("model")["f1_score"].idxmax()]
    .reset_index(drop=True)
)

print("Best Random Forest thresholds by F1-score:")
display(best_rf_thresholds_by_f1.round(4))

# Business-focused subset
business_threshold_view = rf_threshold_comparison_df[[
    "model",
    "threshold",
    "precision",
    "recall",
    "f1_score",
    "false_positives",
    "false_negatives"
]].copy()

display(business_threshold_view.round(4))

# Plot threshold trade-offs for each model
for model_name in rf_threshold_comparison_df["model"].unique():
    plot_df = rf_threshold_comparison_df[
        rf_threshold_comparison_df["model"] == model_name
    ]

    ax = plot_df.plot(
        x="threshold",
        y=["precision", "recall", "f1_score"],
        marker="o",
        figsize=(8, 5)
    )

    ax.set_title(f"Threshold Trade-Off — {model_name}")
    ax.set_xlabel("Decision Threshold")
    ax.set_ylabel("Score")
    ax.set_ylim(0, 1.05)
    plt.tight_layout()
    plt.show()

# Plot false positives and false negatives by threshold
for model_name in rf_threshold_comparison_df["model"].unique():
    plot_df = rf_threshold_comparison_df[
        rf_threshold_comparison_df["model"] == model_name
    ]

    ax = plot_df.plot(
        x="threshold",
        y=["false_positives", "false_negatives"],
        marker="o",
        figsize=(8, 5)
    )

    ax.set_title(f"False Positives vs False Negatives — {model_name}")
    ax.set_xlabel("Decision Threshold")
    ax.set_ylabel("Number of Cases")
    plt.tight_layout()
    plt.show()

print("Random Forest threshold analysis completed successfully.")
```

### 6.16 Business Cost Simulation

Statistical metrics such as accuracy, precision, recall, F1-score, ROC-AUC, and PR-AUC are useful for evaluating model performance. However, credit-risk decisions also require business-cost reasoning.

In this project, the two main prediction errors have different business consequences:

1. **False Negative**
   - The model predicts that a borrower is safe.
   - The borrower actually defaults.
   - This may lead to financial loss because a risky loan is approved.

2. **False Positive**
   - The model predicts that a borrower is risky.
   - The borrower would actually repay.
   - This may create lost revenue opportunity or unnecessary manual review.

Because false negatives are usually more costly in credit-risk settings, this simulation assumes the following relative cost structure:

| Error Type | Assumed Cost |
|---|---:|
| False Negative | 10 cost units |
| False Positive | 1 cost unit |

These are simplified hypothetical values. In a real financial institution, the actual costs would depend on loan amount, interest income, recovery rate, operational review cost, and risk policy.

The goal of this simulation is not to estimate exact financial profit. The goal is to compare model-threshold combinations under a business-oriented cost assumption.

```python
# ============================================================
# Business Cost Simulation
# ============================================================

# Define hypothetical relative costs
COST_FALSE_NEGATIVE = 10
COST_FALSE_POSITIVE = 1

print("Cost assumptions:")
print(f"False Negative cost: {COST_FALSE_NEGATIVE} cost units")
print(f"False Positive cost: {COST_FALSE_POSITIVE} cost unit")

# ------------------------------------------------------------
# Helper function for cost calculation
# ------------------------------------------------------------

def calculate_business_cost(false_positives, false_negatives,
                            cost_fp=COST_FALSE_POSITIVE,
                            cost_fn=COST_FALSE_NEGATIVE):
    """
    Calculate total business cost from false positives and false negatives.
    """
    return (false_positives * cost_fp) + (false_negatives * cost_fn)


# ------------------------------------------------------------
# Candidate 1: Logistic Regression threshold results
# ------------------------------------------------------------

log_reg_cost_candidates = threshold_results_df.copy()
log_reg_cost_candidates["model"] = "Logistic Regression Baseline"

log_reg_cost_candidates = log_reg_cost_candidates[
    log_reg_cost_candidates["threshold"].isin([0.40, 0.50])
].copy()

# ------------------------------------------------------------
# Candidate 2: Random Forest threshold results
# ------------------------------------------------------------

rf_cost_candidates = rf_threshold_comparison_df[
    (
        (rf_threshold_comparison_df["model"] == "Random Forest Baseline") &
        (rf_threshold_comparison_df["threshold"].isin([0.30, 0.40, 0.50]))
    )
    |
    (
        (rf_threshold_comparison_df["model"] == "Random Forest Tuned") &
        (rf_threshold_comparison_df["threshold"].isin([0.35, 0.45, 0.50]))
    )
].copy()

# ------------------------------------------------------------
# Combine candidates
# ------------------------------------------------------------

cost_simulation_df = pd.concat(
    [log_reg_cost_candidates, rf_cost_candidates],
    ignore_index=True
)

# Calculate total business cost
cost_simulation_df["total_business_cost"] = cost_simulation_df.apply(
    lambda row: calculate_business_cost(
        false_positives=row["false_positives"],
        false_negatives=row["false_negatives"]
    ),
    axis=1
)

# Create no-model baseline:
# If all borrowers are approved, all actual defaulters become missed defaulters.
all_approve_false_negatives = int(y_test.sum())
all_approve_false_positives = 0
all_approve_cost = calculate_business_cost(
    false_positives=all_approve_false_positives,
    false_negatives=all_approve_false_negatives
)

# Add savings compared with all-approve baseline
cost_simulation_df["cost_saved_vs_all_approve"] = (
    all_approve_cost - cost_simulation_df["total_business_cost"]
)

cost_simulation_df["cost_reduction_vs_all_approve_pct"] = (
    cost_simulation_df["cost_saved_vs_all_approve"] / all_approve_cost * 100
).round(2)

# Select and order columns
cost_columns = [
    "model",
    "threshold",
    "precision",
    "recall",
    "f1_score",
    "false_positives",
    "false_negatives",
    "total_business_cost",
    "cost_saved_vs_all_approve",
    "cost_reduction_vs_all_approve_pct"
]

cost_simulation_view = cost_simulation_df[cost_columns].sort_values(
    by="total_business_cost",
    ascending=True
)

display(cost_simulation_view.round(4))

print("No-model baseline: approve all borrowers")
print(f"False negatives: {all_approve_false_negatives}")
print(f"False positives: {all_approve_false_positives}")
print(f"Total business cost: {all_approve_cost}")

# Identify lowest-cost model-threshold combination
best_cost_candidate = cost_simulation_view.iloc[0]

print("\nLowest-cost model-threshold combination:")
display(pd.DataFrame([best_cost_candidate]).round(4))

# Plot business cost comparison
ax = cost_simulation_view.set_index(
    cost_simulation_view["model"] + " | Threshold " + cost_simulation_view["threshold"].astype(str)
)["total_business_cost"].plot(
    kind="bar",
    figsize=(10, 5)
)

ax.set_title("Business Cost Simulation by Model and Threshold")
ax.set_xlabel("Model and Threshold")
ax.set_ylabel("Total Business Cost")
ax.tick_params(axis="x", rotation=45)
plt.tight_layout()
plt.show()

print("Business cost simulation completed successfully.")
```

### 7.1 Objective

After selecting the final model, it is important to understand which features contribute most to the model's predictions.

```python
# ============================================================
# Model Interpretation: Feature Importance
# ============================================================

# Extract fitted preprocessing and model steps from the final selected model
final_preprocessor = final_model.named_steps["preprocessor"]
final_rf_classifier = final_model.named_steps["model"]

# Extract encoded categorical feature names
encoded_categorical_features = (
    final_preprocessor
    .named_transformers_["cat"]
    .get_feature_names_out(categorical_cols)
)

# Combine numerical and encoded categorical feature names
encoded_feature_names = numerical_cols + list(encoded_categorical_features)

# Extract Random Forest feature importances
feature_importance_df = pd.DataFrame({
    "encoded_feature": encoded_feature_names,
    "importance": final_rf_classifier.feature_importances_
})

feature_importance_df = feature_importance_df.sort_values(
    by="importance",
    ascending=False
).reset_index(drop=True)

# Display top encoded features
top_encoded_features = feature_importance_df.head(15)
display(top_encoded_features)

# Plot top encoded feature importances
plt.figure(figsize=(9, 6))
sns.barplot(
    data=top_encoded_features,
    x="importance",
    y="encoded_feature"
)

plt.title("Top 15 Encoded Feature Importances — Final Random Forest Model")
plt.xlabel("Importance")
plt.ylabel("Encoded Feature")
plt.tight_layout()
plt.show()


# ============================================================
# Aggregate encoded feature importances back to original features
# ============================================================

def map_to_original_feature(encoded_feature, original_features):
    """
    Map an encoded feature name back to its original feature.
    """
    for original_feature in original_features:
        if encoded_feature == original_feature:
            return original_feature
        if encoded_feature.startswith(original_feature + "_"):
            return original_feature
    return encoded_feature


original_feature_list = numerical_cols + categorical_cols

feature_importance_df["original_feature"] = feature_importance_df["encoded_feature"].apply(
    lambda x: map_to_original_feature(x, original_feature_list)
)

aggregated_feature_importance = (
    feature_importance_df
    .groupby("original_feature", as_index=False)["importance"]
    .sum()
    .sort_values(by="importance", ascending=False)
    .reset_index(drop=True)
)

display(aggregated_feature_importance)

# Plot aggregated feature importances
plt.figure(figsize=(9, 6))
sns.barplot(
    data=aggregated_feature_importance,
    x="importance",
    y="original_feature"
)

plt.title("Aggregated Feature Importance by Original Feature")
plt.xlabel("Aggregated Importance")
plt.ylabel("Original Feature")
plt.tight_layout()
plt.show()

print("Model interpretation completed successfully.")
```

```python
# ============================================================
# Responsible AI Diagnostic: Subgroup Performance
# ============================================================

# Create test-set predictions for the final selected model
final_test_proba = final_model.predict_proba(X_test)[:, 1]
final_test_pred = (final_test_proba >= final_threshold).astype(int)

# Combine test features, true labels, predictions, and probabilities
responsible_ai_test_df = X_test.copy()
responsible_ai_test_df["y_true"] = y_test.values
responsible_ai_test_df["y_pred"] = final_test_pred
responsible_ai_test_df["default_probability"] = final_test_proba


def subgroup_performance_table(data, group_col):
    """
    Calculate subgroup-level performance metrics for a categorical feature.
    """
    subgroup_results = []

    for group_value, group_data in data.groupby(group_col):
        y_true_group = group_data["y_true"]
        y_pred_group = group_data["y_pred"]

        # Skip very small groups to avoid unstable metrics
        if len(group_data) < 30:
            continue

        # Use labels=[0, 1] to make the confusion matrix robust
        tn, fp, fn, tp = confusion_matrix(
            y_true_group,
            y_pred_group,
            labels=[0, 1]
        ).ravel()

        subgroup_results.append({
            "group_feature": group_col,
            "group_value": group_value,
            "group_size": len(group_data),
            "actual_default_rate": y_true_group.mean(),
            "predicted_default_rate": y_pred_group.mean(),
            "precision": precision_score(y_true_group, y_pred_group, zero_division=0),
            "recall": recall_score(y_true_group, y_pred_group, zero_division=0),
            "f1_score": f1_score(y_true_group, y_pred_group, zero_division=0),
            "false_positives": fp,
            "false_negatives": fn,
            "true_positives": tp,
            "true_negatives": tn
        })

    return pd.DataFrame(subgroup_results)


# Subgroup checks for available categorical variables
subgroup_features = [
    "person_home_ownership",
    "loan_intent",
    "loan_grade",
    "cb_person_default_on_file"
]

subgroup_performance_results = []

for feature in subgroup_features:
    subgroup_table = subgroup_performance_table(
        responsible_ai_test_df,
        group_col=feature
    )
    subgroup_performance_results.append(subgroup_table)

subgroup_performance_df = pd.concat(
    subgroup_performance_results,
    ignore_index=True
)

# Format selected rate columns
subgroup_display = subgroup_performance_df.copy()
subgroup_display["actual_default_rate"] = (
    subgroup_display["actual_default_rate"] * 100
).round(2)

subgroup_display["predicted_default_rate"] = (
    subgroup_display["predicted_default_rate"] * 100
).round(2)

display(subgroup_display.round(4))

print("Responsible AI subgroup diagnostic completed successfully.")
print("Important: This is not a full fairness audit because protected attributes are not available.")
print("The table should be used to identify groups where model performance may require closer review.")
```
